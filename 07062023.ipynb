### PLOTTING WITH PANDAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./Car_sales (3).csv')
data.head(10)
```

```python
data.rename(columns={'__year_resale_value':'year_resale_value'},inplace=True)
```

###### Dropping na values

```python
data.dropna(inplace=True)
```

```python
data.info()
```

### Plots on numerical value coloumns

#### Histogram

```python
data.Horsepower.plot.hist(figsize=(6,3), cmap='brg');
```

#### Box plot

```python
data.Sales_in_thousands.plot.box(figsize=(4,3),cmap='gist_earth');
```

```python
data.Horsepower.plot.box(figsize=(4,3),cmap='Blues_r');
```

#### Subplots

```python
data.plot.box(figsize=(22,4),subplots=True);
```

```python
data.plot.hist(figsize=(6,18),subplots=True);
```

#### Line chart

```python
data.sort_values(by='year_resale_value').plot.line('year_resale_value', 'Price_in_thousands',figsize=(6,3),cmap='inferno');
```

#### BAR PLOT

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2),cmap='turbo_r');
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3));
```

#### Area chart

```python
data.plot.area('Manufacturer','Price_in_thousands',figsize=(6,3),cmap='gray');
```

#### HORIZONTAL BAR CHART

```python
data.sort_values(by='Price_in_thousands')[:5].plot.barh('Model','Power_perf_factor',figsize=(6,4));
```

#### Scatter plot

```python
data.sort_values(by='Price_in_thousands')[:5].plot.scatter('Latest_Launch','Price_in_thousands',figsize=(10,3));
```

```python
data.plot.box(figsize=(22,5));
```

```python
data.sort_values(by='Engine_size',ascending=False)[:10].Engine_size.plot.pie();
```

#### KDE plot

```python
data.plot.kde(figsize=(10,3.8));
```

#### subplots

```python
data.plot.kde(figsize=(10,12),subplots=True);
```

#### Hexbin plot

```python
data.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15,cmap='autumn');
```
